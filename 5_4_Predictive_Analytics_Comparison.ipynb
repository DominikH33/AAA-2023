```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**reading evaluation dataframes**

```python
file_path = "./data/datasets/"

summary_svm_census_24H = pd.read_pickle(f"{file_path}summary_svm_census_24H.pkl")
summary_svm_census_6H = pd.read_pickle(f"{file_path}summary_svm_census_6H.pkl")
summary_svm_7_24H = pd.read_pickle(f"{file_path}summary_svm_7_24H.pkl")
summary_svm_7_6H = pd.read_pickle(f"{file_path}summary_svm_7_6H.pkl")
summary_svm_6_24H = pd.read_pickle(f"{file_path}summary_svm_6_24H.pkl")
summary_svm_6_6H = pd.read_pickle(f"{file_path}summary_svm_6_6H.pkl")

summary_nn_7_24H = pd.read_pickle(f"{file_path}summary_nn_7_24H.pkl")
summary_nn_7_6H = pd.read_pickle(f"{file_path}summary_nn_7_6H.pkl")
summary_nn_6_24H = pd.read_pickle(f"{file_path}summary_nn_6_24H.pkl")
summary_nn_6_6H = pd.read_pickle(f"{file_path}summary_nn_6_6H.pkl")
```

```python
# Create comparison plot
def plot_comparison(df1, df2, title):
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))
    
    ax1 = df1.iloc[[1, 3, 4, 5, 6], [0, 1, 3, 4]].plot(kind='bar', ax=axes[0], color=['blue', 'orange', 'green', 'red'])
    ax1.set_title('24-Hour Resolution')
    ax1.set_ylabel('Values')
    y_max1 = df1.iloc[[1, 3, 4, 5, 6], [0, 1, 3, 4]].max().max()
    ax1.set_ylim([0, y_max1])
    
    ax2 = df2.iloc[[1, 3, 4, 5, 6], [0, 1, 3, 4]].plot(kind='bar', ax=axes[1], color=['blue', 'orange', 'green', 'red'])
    ax2.set_title('6-Hour Resolution')
    ax2.set_ylabel('Error')
    y_max2 = df2.iloc[[1, 3, 4, 5, 6], [0, 1, 3, 4]].max().max()
    ax2.set_ylim([0, max(y_max1, y_max2)])
    
    plt.suptitle(title)
    plt.tight_layout()
    plt.show()
```

## SVM Spatio-Temporal Resolution Comparison

### Census Tract Resolution

```python
plot_comparison(summary_svm_census_24H, summary_svm_census_6H, 'SVM Census Tract Resolution')
```

### Hexagon Resolution 7

```python
plot_comparison(summary_svm_7_24H, summary_svm_7_6H, 'SVM Hexagon Resolution 7')
```

### Hexagon Resolution 6

```python
plot_comparison(summary_svm_6_24H, summary_svm_6_6H, 'SVM Hexagon Resolution 6')
```

## NN Spatio-Temporal Resolution Comparison

### Hexagon Resolution 7

```python
plot_comparison(summary_nn_7_24H, summary_nn_7_6H, 'NN Hexagon Resolution 7')
```

### Hexagon Resolution 6

```python
plot_comparison(summary_nn_6_24H, summary_nn_6_6H, 'NN Hexagon Resolution 6')
```

## SVM vs NN

```python
# Create comparison plot
def plot_comparison2(df1, df2, title1, title2, title):
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))
    
    ax1 = df1.iloc[[1, 3, 4, 5, 6], [0, 1, 3, 4]].plot(kind='bar', ax=axes[0], color=['blue', 'orange', 'green', 'red'])
    ax1.set_title(title1)
    ax1.set_ylabel('Values')
    y_max1 = df1.iloc[[1, 3, 4, 5, 6], [0, 1, 3, 4]].max().max()
    ax1.set_ylim([0, y_max1])
    
    ax2 = df2.iloc[[1, 3, 4, 5, 6], [0, 1, 3, 4]].plot(kind='bar', ax=axes[1], color=['blue', 'orange', 'green', 'red'])
    ax2.set_title(title2)
    ax2.set_ylabel('Error')
    y_max2 = df2.iloc[[1, 3, 4, 5, 6], [0, 1, 3, 4]].max().max()
    ax2.set_ylim([0, max(y_max1, y_max2)])
    
    plt.suptitle(title)
    plt.tight_layout()
    plt.show()
```

### Hexagon Resolution 7

```python
plot_comparison2(summary_svm_7_24H, summary_nn_7_24H, 'SVM 24H Resolution', 'NN 24H Resolution', 'SVM vs NN : Hexagon Resolution 7')
```

```python
plot_comparison2(summary_svm_7_6H, summary_nn_7_6H, 'SVM 6H Resolution', 'NN 6H Resolution', 'SVM vs NN : Hexagon Resolution 7')
```
